```python
import requests
url = 'https://www.buffalo.edu/class-schedule?switch=showcourses&semester=fall&division=UGRD&dept=ASL'
s = requests.get(url).text
```

```python
import bs4
```

```python
b = bs4.BeautifulSoup(s,'lxml')
type(b)
```

```python
for item in b:
    print(item.name)
```

```python
for item in b.html:
    print(item.name)
```

```python
for item in b.html.body:
    if item.name:
        print(item.name)
```

```python
count = 0
wanted = 3
for item in b.html.body:
    if item.name=='table':
        count += 1
        if count==wanted:
            print(item.name)
```

```python
tables = b.find_all('table')
mytable = tables[5]
```

```python
for row in mytable:
    if row.name:
        tds = row.find_all('td')
        if len(tds)==11:
            print(tds[7].text.strip())
            try:
                print(row['onclick'])
            except:
                pass
```
