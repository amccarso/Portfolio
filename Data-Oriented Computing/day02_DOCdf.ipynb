```python
def numberofvowels(string):
    count = 0
    vowels = 'aeiou'
    for letter in string:
        if letter in vowels:
            count += 1
    return count
```

```python
def reversalinlist(string, df, x):
    for i in range(len(df['phrase'])):
        if df['phrase'][i] == string[::-1] and x != i:
            return True
    return False
```

```python
pd.options.mode.chained_assignment = None
df['length'] = 0
df['Capitalized'] = 0
df['number of vowels'] = 0
df['proportion of vowels'] = 0
df['reversal'] = 0
df['isPalindrome'] = 0
df['alphabetical order'] = 0
df['reversal in list'] = 0
for x in range(len(df['phrase'])):
    df['length'][x] = len(df['phrase'][x])
    df['Capitalized'][x] = df['phrase'][x][0].isupper()
    df['number of vowels'][x] = numberofvowels(df['phrase'][x])
    df['proportion of vowels'][x] = df['length'][x]/df['number of vowels'][x]
    df['reversal'][x] = df['phrase'][x][::-1]
    df['isPalindrome'][x] = df['phrase'][x] == df['reversal'][x]
    df['alphabetical order'][x] = sorted(df['phrase'][x])
    df['reversal in list'][x] = reversalinlist(df['phrase'][x],df,x)
df
```
