```python
ca_babies = state_baby_names.loc[state_baby_names['State'] == 'CA', :]
```

```python
ca_babies.groupby(['Year', 'Gender'])
```

```python
def popular(s):
    """Receives s, a Pandas series, of baby names sorted from highest count to lowest count.
    Returns the most popular name in s"""
    
    return s.iloc[0]
```

```python
sorted_ca_babies = ca_babies.sort_values('Count', ascending = False).groupby(['Year', 'Gender']).agg(popular)
```

```python
sorted_ca_babies
```
